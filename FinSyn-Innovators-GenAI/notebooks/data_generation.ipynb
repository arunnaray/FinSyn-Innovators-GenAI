```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/employee_data_reference.csv')
df
```

```python
df.rename(columns={'Years of experience' : 'Years of Experience'}, inplace=True)
```

```python
import numpy as np

# Update ID column to the format EMPXX
df['ID'] = df['ID'].apply(lambda x: f'EMP{x:02d}')

# Add random Band Level
df['Band Level'] = np.random.choice([28, 30, 35, 40, 45, 60, 70, 90, 200], size=len(df))

# Add random Laptop Assigned
df['Laptop Assigned'] = np.random.choice(['MacBook Air', 'MacBook Pro', 'Dell', 'Lenevo', 'Windows Surface'], size=len(df))

# Add Years of Experience based on Salary (assuming higher salary implies more experience)
df['Years of experience'] = df['Salary'].apply(lambda x: np.random.randint(10, 20) if x > 75000 
                                               else np.random.randint(2, 10))

# Add random Number of tickets raised in a month
df['Number of tickets raised in a month'] = np.random.randint(1, 6, size=len(df))

# Add random Last Performance Rating
df['Last Performance Rating'] = np.random.randint(1, 6, size=len(df))

# Display the updated dataset
df.head()
```

```python
df.columns
```

```python
df = df[['ID', 'Name', 'Age', 'Department', 'Salary', 'Date_of_Joining',
       'Band Level', 'Years of experience', 'Last Performance Rating',
       'Laptop Assigned', 'Number of tickets raised in a month', ]]
```
